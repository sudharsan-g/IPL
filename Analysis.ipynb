```python
import pandas as pd
import numpy as np 
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
mega_df = pd.read_csv("./Dataset/mega_table.csv")
df_2023 = pd.read_excel("./Dataset/Book 2.xlsx")
mega_df
```

## Total_runs over Seasons

```python
teams_23 = df_2023["Team"].unique()
i = 0
total_run_of_all_teams_ss = pd.DataFrame(columns=list(mega_df["Season"].unique()),index = teams_23)
fig ,ax = plt.subplots(5,2,figsize=(15,23))
for row in range(5):
    for column in range(2):
        total_run_of_all_teams_ss.loc[teams_23[i]] = mega_df[mega_df["BattingTeam"]==teams_23[i]].groupby("Season")["total_run"].sum()
        mega_df[mega_df["BattingTeam"]==teams_23[i]].groupby("Season")["total_run"].sum().plot(kind="line",marker="o",ax=ax[row,column])
        ax[row,column].title.set_text(teams_23[i])
        i+=1
```

```python
team_specific_df =mega_df[["ID","Team1","Team2","City","TossWinner","TossDecision","WinningTeam","Season","MatchNumber","WonBy"]].drop_duplicates(subset="ID")
```

## Season over Team winning over Chase=`Wickets` or Puton=`Runs`

```python
seasons = mega_df["Season"].unique()
i = 0
winning_teams_over_season = team_specific_df[["WinningTeam","WonBy","Season"]].groupby("Season").value_counts().unstack()
fig ,ax = plt.subplots(8,2,figsize=(10,23))
fig.tight_layout(w_pad=20.0)
for row in range(8):
    for column in range(2):
        if i !=15:
            season_by_win = winning_teams_over_season[team_specific_df[["WinningTeam","WonBy","Season"]].groupby("Season").value_counts().unstack().index.get_level_values(0)==seasons[i]]
            season_by_win.plot(kind="barh",ax=ax[row,column])
        i+=1
```
