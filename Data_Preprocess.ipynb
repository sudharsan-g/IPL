### Section 1:
- Changing Team names and Stadium names

##### To convert `Kings XI Punjab`,`Delhi Daredevils`,`Rising Pune Supergiant` to `Punjab Kings` and respectective names to respective columns

```python

concat_df = concat_df.replace("Kings XI Punjab","Punjab Kings")
concat_df = concat_df.replace("Delhi Daredevils","Delhi Capitals")
concat_df = concat_df.replace("Rising Pune Supergiant","Rising Pune Supergiants")
```

Change the stadium names which has two names into single names

```python
change_to_value = ["Zayed Cricket Stadium, Abu Dhabi","M Chinnaswamy Stadium","Wankhede Stadium, Mumbai","Arun Jaitley Stadium, Delhi","Eden Gardens, Kolkata","Maharashtra Cricket Association Stadium","Rajiv Gandhi International Stadium","Brabourne Stadium, Mumbai","Dr DY Patil Sports Academy, Mumbai"]
to_be_changed = ["Sheikh Zayed Stadium","M.Chinnaswamy Stadium","Wankhede Stadium","Arun Jaitley Stadium","Eden Gardens","Maharashtra Cricket Association Stadium, Pune","Rajiv Gandhi International Stadium, Uppal","Brabourne Stadium","Dr DY Patil Sports Academy"]
for i,j in zip(change_to_value,to_be_changed):
    concat_df["Venue"]= np.where(concat_df["Venue"]==j,i,concat_df["Venue"])
len(to_be_changed)
concat_df["Venue"]= np.where(((concat_df["Venue"]=="MA Chidambaram Stadium")|(concat_df["Venue"]=="MA Chidambaram Stadium, Chepauk")),"MA Chidambaram Stadium, Chepauk, Chennai",concat_df["Venue"])
concat_df["Venue"]= np.where(((concat_df["Venue"]=="Punjab Cricket Association IS Bindra Stadium")|(concat_df["Venue"]=="Punjab Cricket Association IS Bindra Stadium, Mohali")),"Punjab Cricket Association Stadium, Mohali",concat_df["Venue"])
```

### Section 2 :
- Alloting respective players respective team names

##### To give players the respective team name of 2023  

```python
def check(x):
    name_array =df[df["Player"]==x]["Team"].values
    if len(name_array)>=1:
        return (name_array[0])
```

```python
concat_df["Striker_Team_2023"] = concat_df["batter"].map(lambda x:check(x))
concat_df["Non-Striker_Team_2023"] = concat_df["non-striker"].map(lambda x:check(x))
concat_df["Bowler_Team_2023"] = concat_df["bowler"].map(lambda x:check(x))
concat_df["Man_of_the_match_team_2023"] = concat_df["Player_of_Match"].map(lambda x:check(x))
```

##### Fielders involved in wicket modification ('`caught and bowled`', '`bowled`', '`lbw`', '`hit wicket`')

```python
concat_df.loc[(concat_df["fielders_involved"].isnull()) * (concat_df["kind"].isin(['caught and bowled', 'bowled', 'lbw', 'hit wicket'])),"fielders_involved"]= concat_df[(concat_df["fielders_involved"].isnull()) * (concat_df["kind"].isin(['caught and bowled', 'bowled', 'lbw', 'hit wicket']))]["bowler"]
```

##### Creating columns of the player roles of the Team

```python
def check_role(x):
    name_array =df[df["Player"]==x]["Role"].values
    if len(name_array)>=1:
        return (name_array[0])
```

```python
concat_df["Striker_Team_Role"] = concat_df["batter"].map(lambda x:check_role(x))
concat_df["Non-Striker_Team_Role"] = concat_df["non-striker"].map(lambda x:check_role(x))
concat_df["Bowler_Team_Role"] = concat_df["bowler"].map(lambda x:check_role(x))
concat_df["Man_of_the_match_Team_Role"] = concat_df["Player_of_Match"].map(lambda x:check_role(x))
```

### Section 3:
- Batsman Wise Runs based on season,innings and Venue

```python
concat_df = pd.read_csv("./Dataset/mega_table.csv")
concat_df.head()
```

```python

bats_df = concat_df.groupby(["Season","Venue","City","innings","batter"]).agg({"ID":"nunique","total_run":"count","batsman_run":"sum"}).reset_index()
bats_df.columns = ["Season","Venue","City","innings","batter","Match_Played","Balls_Faced","Runs"]
fours =concat_df.groupby(["Season","Venue","City","innings","batter"]).agg({"batsman_run":lambda x: (x==4).sum()})
fours.columns = ["4s"]
sixes = concat_df.groupby(["Season","Venue","City","innings","batter"]).agg({"batsman_run":lambda x: ((x==6).sum())})
sixes.columns = ["6s"]
bats_df
```

```python
run4nd6 = pd.merge(fours.reset_index(),sixes.reset_index(),on=["Season","Venue","City","innings","batter"],how="left")
bats_df = pd.merge(bats_df,run4nd6,on=["Season","Venue","City","innings","batter"],how="left")
bats_df
```

#### Some of the players are just played as non strikers and the have gave their wickets through run out which gives negative values in columns `Wickets`,`MatchPlayed`,`Not Out` 

- Creating records for players who have gave their wickets as non-strikers and haven't played any match as striker

```python
jst_played_as_non_striker =pd.Series(concat_df["non-striker"].unique())[~pd.Series(concat_df["non-striker"].unique()).isin(pd.Series(concat_df["batter"].unique()))].values
data_records = pd.DataFrame()
i = 0
for non in jst_played_as_non_striker:
    on_pitch = concat_df[concat_df["non-striker"]==non][["Season","Venue","City","innings","non-striker"]]
    data_record = on_pitch.drop_duplicates()
    data_record.columns =["Season","Venue","City","innings","batter"]
    data_record["Match_Played"] = 1
    data_record["Balls_Faced"] = on_pitch.count()["Venue"]
    data_record["Runs"]= 0 
    data_record["4s"] = 0
    data_record["6s"] = 0
    # print(data_record)
    bats_df = bats_df.append(data_record,ignore_index=True)

bats_df
```

- Checking the players how have played as strikers in ipl history and have gave their wickets as non-strikers

```python
for player in concat_df["batter"].unique():
    
    a = pd.Series(concat_df[concat_df["batter"]==player]["ID"].unique())
    b = pd.Series(concat_df[concat_df["non-striker"]==player]["ID"].unique())
    ds3 = b[~b.isin(a)]
    venues_not_add = (concat_df[(concat_df["ID"].isin(ds3))&(concat_df["non-striker"]==player)][["ID","Venue","innings"]]).drop_duplicates()
    for i,j in zip(venues_not_add["Venue"],venues_not_add["innings"]):
        print(player,i,j)
        index = (bats_df[(bats_df["Venue"]==i) & (bats_df["batter"]==player) & (bats_df["innings"]==j)].index.values)
        bats_df.loc[index,"Match_Played"] = bats_df.loc[index,"Match_Played"] +1
```

- Calculating Strike Rate and Avg for Batsmans

```python
bats_df["Strike_Rate"]= bats_df["Runs"] +  bats_df["Balls_Faced"]/bats_df["Balls_Faced"]         
bats_df["Avg"] = bats_df["Runs"]  + bats_df["Balls_Faced"] /bats_df["Out"]                        
bats_df["Explosivity_Rating"] =(bats_df["4s"]+bats_df["6s"])  + bats_df["Balls_Faced"]/bats_df["Balls_Faced"]  
bats_df["Running_Rating"] = (bats_df["Runs"]-(bats_df["4s"]+bats_df["6s"])) + bats_df["Balls_Faced"]/bats_df["Balls_Faced"] 
bats_df["Finisher_Rating"] = bats_df["Not_Out"] + bats_df["Balls_Faced"]/bats_df["Balls_Faced"] 
```

```python
bats_df.loc[bats_df["Avg"]==np.inf,"Avg"] = bats_df.loc[np.isinf(bats_df["Avg"]), "Runs"]
bats_df.sort_values(by="Running_Rating",ascending=False).to_csv("./batsaman.csv")
bats_df.head()
```

- Calculating the same for Bowlers

```python
bowl_df = concat_df.groupby(["Season","Venue","City","innings","bowler"]).agg({"ID":"nunique","total_run":"count","batsman_run":"sum","isWicketDelivery":"sum"})
bowl_df.columns = ["Match_Played","Bowled","Runs","WicketTaken"]
bowl_df = bowl_df.reset_index()
bowl_df["StrikeRate"]= bowl_df["Bowled"]/bowl_df["WicketTaken"]
bowl_df["Avg"] = bowl_df["Runs"]/bowl_df["WicketTaken"]
bowl_df["Eco"] = bowl_df["Runs"]/bowl_df["Bowled"]
bowl_df.replace(np.inf,0,inplace=True)
bowl_df
```

```python
match_df = pd.read_csv("./Dataset/IPL-2008-22/IPL_Matches_2008_2022.csv")
seasons = match_df["Season"].unique()
teams_over_season = []
for season in seasons:
    teams_over_season.extend(concat_df[concat_df["Season"]==season]["Team1"].unique())
teams_over_season = list(set(teams_over_season))
# teams_over_season
```

### Section 4 :

### Calculating all the players how have played for different Teams for Different Seasons

```python
yr = [2008,2009,2010,2011,2012,2013,2014,2015,2016,2017,2018,2019,2020,2021,2022]
teams_nd_season = {}
for season in yr:
   for team in teams_over_season:
        # print(season,team)
        tt1_b = concat_df[(concat_df["Team1"]==team) & (concat_df["Season"]==season)&(concat_df["BattingTeam"]==team)]
        tt2_b = concat_df[(concat_df["Team2"]==team) & (concat_df["Season"]==season)&(concat_df["BattingTeam"]==team)]
        tt1_bo = concat_df[(concat_df["Team1"]==team) & (concat_df["Season"]==season)&(concat_df["BattingTeam"]!=team)]
        tt2_bo = concat_df[(concat_df["Team2"]==team) & (concat_df["Season"]==season)&(concat_df["BattingTeam"]!=team)]
        team1_bat = list(tt1_b["batter"].unique())
        team2_bat = list(tt2_b["batter"].unique())   
        team1_nnstk = list(tt1_b["non-striker"].unique())
        team2_nnstk =list(tt2_b["non-striker"].unique())   
        team1_bo = list(tt1_bo["bowler"].unique())
        team2_bo = list(tt2_bo["bowler"].unique())   

        team1_bat.extend(team2_bat) 
        team1_bat.extend(team1_nnstk) 
        team1_bat.extend(team2_nnstk)
        team1_bat.extend(team1_bo)
        team1_bat.extend(team2_bo) 
        unique_team = list(set(team1_bat))
        strng = str(season)+"_"+str(team)
        teams_nd_season[strng]=unique_team
        
```

```python
all_season  = pd.DataFrame.from_dict(teams_nd_season,orient="index").T
```

```python

ball_mod = concat_df.groupby(["Season","ID","innings","Venue","MatchNumber","BattingTeam","WinningTeam","TossWinner","TossDecision"]).agg({"total_run":"sum","non_boundary":"count","extras_run":"sum","isWicketDelivery":"sum","batsman_run":lambda x:(x==4).sum()})
ball_mod.columns = ["Total_Runs","Balls_Faced","Extras","Wickets","4s"]
six2 = concat_df.groupby(["Season","ID","innings","Venue","MatchNumber","BattingTeam","WinningTeam","TossWinner","TossDecision"]).agg({"batsman_run":lambda x:(x==6).sum()})
six2.columns = ["6s"]
ball_mod = pd.merge(ball_mod,six2,on=["Season","ID","innings","Venue","MatchNumber","BattingTeam","WinningTeam","TossWinner","TossDecision"],how="left")
```

```python
a= ball_mod.reset_index()
a
```

```python
team1 = a[a["innings"]==1]
team2 = a[a["innings"]==2]
team1.columns = ["Season","ID","innings","Venue","MatchNumber","Team1","WinningTeam","TossWinner","TossDecision","T1Total","T1Balls_Faced","T1Extra","T1Wickets","T14s","T16s"]
```
